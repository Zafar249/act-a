```python
from pycoingecko import CoinGeckoAPI
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
cg = CoinGeckoAPI()
```

```python
# Testing the connection to see whether it is stable
cg.ping()
```

```python
# Get the price of bitcoin and ehtereum in usdt
price = cg.get_price(ids=["bitcoin","ethereum"], vs_currencies="usd", include_market_cap="true",
        include_24hr_vol="true", include_24hr_change="true", include_last_updated_at="true")
```

```python
price["ethereum"]
```

```python
# Get a list of all coins, id, names and symbols
coins_list = cg.get_coins_list()
```

```python
# Gets data about the coin over the last 7 days
historial_data = cg.get_coin_market_chart_by_id(
    id="ethereum",
    vs_currency="usd",
    days = "7"
)
```
